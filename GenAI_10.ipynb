```python
import torch
from diffusers import Lumina2Text2ImgPipeline
from PIL import Image

# Load the Lumina-Image-2.0 model
model_id = "Alpha-VLLM/Lumina-Image-2.0"
pipe = Lumina2Text2ImgPipeline.from_pretrained(model_id, torch_dtype=torch.bfloat16)
pipe.enable_model_cpu_offload()  # Offload model to CPU to save VRAM; remove if you have sufficient GPU memory

# Define your prompt
prompt = "A serene photograph capturing the golden reflection of the sun on a vast expanse of water."

# Generate the image
image = pipe(prompt).images[0]

# Save the image
image.save("lumina_generated_image.png")
```
