```python
from transformers import pipeline

def qa_chatbot():
    qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

    context = """
    Artificial Intelligence (AI) is a branch of computer science that aims to create intelligent machines.
    It has become an essential part of the technology industry. AI research deals with the problem of creating
    machines capable of performing tasks that require human intelligence, such as visual perception, speech
    recognition, decision-making, and language translation.
    """

    print("Chatbot initialized! Ask a question or type 'exit' to quit.")

    while True:
        question = input("You: ")

        if question.lower() == "exit":
            print("Chatbot: Goodbye!")
            break

        answer = qa_pipeline(question=question, context=context)
        print(f"Chatbot: {answer['answer']}")

if __name__ == "__main__":
    qa_chatbot()
```
